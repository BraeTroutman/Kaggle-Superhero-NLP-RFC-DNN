```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier 
```

```python
data = pd.read_csv('superheroes_nlp_dataset.csv')
data.head()
```

```python
data = data.drop(data.columns.drop('creator').drop('history_text'), axis=1)
data
```

```python
def classer (x):
    if x == 'Marvel Comics':
        return int(0)
    elif x == 'DC Comics':
        return int(1)
data['creator'] = data['creator'].apply(classer)
data
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X = data['history_text'].values
Y = data['creator'].values

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.preprocessing import StandardScaler
```

```python
cv = CountVectorizer()
x_train = cv.fit_transform(x_train)
x_test = cv.transform(x_test)
```

```python
tf = TfidfTransformer()
x_train = tf.fit_transform(x_train)
x_test = tf.transform(x_test)

x_train = x_train.toarray()
x_test = x_test.toarray()
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
rfc = RandomForestClassifier(n_jobs=-1, n_estimators=200)
rfc.fit(x_train, y_train)
```

```python
rfc.score(x_test, y_test)
```
